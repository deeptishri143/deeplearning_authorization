```
data= pd.read_csv('/content/generated_from_DP_Reg_model_sample_03_conv_100.csv')
data
```

```
category = data.select_dtypes(include=['category','object','bool'])

cols= category.columns
```

```
category= pd.get_dummies(category)
```

```
category.apply(pd.Series.value_counts)
```

```
numerical = data.select_dtypes(include=['int', 'int64', 'float'])
```

```
Y= numerical.iloc[:,0]
Y.value_counts()
```

```
category.drop(['principal_pob_possession_OTH', 'principal_pob_otherbus_BARTENDING TEACHING ACADMY', 'cob_SCT', 'principal_pob_otherbus_CUSTOMERS SPECIFIED FABRICATION JOBWORK', 'principal_pob_otherbus_None', 	'principal_pob_otherbus_RESTAURANT',	'principal_pob_otherbus_Sale of bakery products',	'principal_pob_otherbus_Tours and Travels',	'principal_pob_otherbus_mfg',	'cob_FCO',	'cob_HUF',	'cob_OTH', 'cob_PUB','cob_PUB', 'oth_rsn_registration_BANK PURPOSE', 'oth_rsn_registration_BUSINESS STARTED',	'oth_rsn_registration_BUSSINESS PURPOSE',	'oth_rsn_registration_JOB WORKS',	'oth_rsn_registration_MAN POWER SOLUTIONS',	'oth_rsn_registration_NEW BUSINESS', 'oth_rsn_registration_OTHER',	'oth_rsn_registration_PURCHASE AND SALE OF CLOTH, CUTTING WASTE', 'oth_rsn_registration_Retail Sales and Service',	'oth_rsn_registration_SALE &SERVICE OF AIR CONDITIONER ELECTRIC ELECTRON'	,'oth_rsn_registration_VOLUNTARY BASIS',	'oth_rsn_registration_mfg of-nut bolt shaft',	'oth_rsn_registration_work contract and supplier'	,'State_sens_commodity_N'], axis =1, inplace= True)
```

```
category.apply(pd.Series.value_counts)
```

```
numerical = numerical.drop(['class_label_sample'], axis=1) # drop label
numerical 
```

```
numerical= numerical.iloc[:, [0,1,2,3,5,52]]
numerical
```

```
df = pd.concat([category, numerical], axis=1)
len(df.columns)
```

```
from sklearn.preprocessing import StandardScaler
cols= df.columns
scaler = StandardScaler() 
df[cols] = scaler.fit_transform(df[cols])
```

```
X= df.iloc[:,1:]
X
```

```
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

```
ann= tf.keras.models.Sequential()
```

```
ann.add(tf.keras.layers.Dense(units= 16, activation= 'relu'))
```

```
ann.add(tf.keras.layers.Dense(units= 8, activation= 'relu'))
```

```
ann.add(tf.keras.layers.Dense(units= 8, activation= 'relu'))
```

```
ann.add(tf.keras.layers.Dense( units=1, activation= "sigmoid"))
```

```
ann.compile(optimizer= 'adam', loss='binary_crossentropy', metrics= ['accuracy'])
```

```
ann.fit(X_train, y_train, batch_size= 32, epochs= 100)
```

```
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
```

```
y_pred= ann.predict(X_test)
y_pred= (y_pred>0.5)
```

```
print ('TEST RESULT')
print ('Accuracy RF:', accuracy_score(y_test, y_pred))
print ('F1 score:', f1_score(y_test,y_pred ,average='weighted'))
print ('Recall:', recall_score(y_test,y_pred ,average='weighted'))
print ('Precision:', precision_score(y_test, y_pred,average='weighted') )
cm_test = metrics.confusion_matrix(y_test, y_pred)
print('Confusion matrix:\n', cm_test)
```
